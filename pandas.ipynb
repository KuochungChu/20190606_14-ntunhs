```python
import pandas as pd

lst=[1,3,45,6,7,6,4,32,23,6]

date=pd.date_range('20190604',periods=10)
s=pd.Series(lst,index=date)

#搭配type方法，觀察資料型態
print(s)
```

```python
#index為日期
#loc 直接用index的值進行查詢

print(s.loc['20190604'])
print(s.loc['20190604':'20190610'])
```

```python
#由0開始算
#loc 直接用index的順序進行查詢

print(s.iloc[6]) 
print(s.iloc[5:7])

print(s.max())
print(s.min()) 
print(s.mean())
print(s.std()) 
```

```python
date=pd.date_range('20190604',periods=3)
print(date)

#將datetimeIndex格式轉為list
date_string=date.strftime("%Y-%m-%d %H:%M:%S").tolist()
print(date_string)

s1=pd.Series([1,2,3],index=date)
s2=pd.Series([4,5,6],index=date)
s3=pd.Series([7,8,9],index=date)

#由series組合為dataframe
df=pd.DataFrame()
df=pd.DataFrame({'c1':s1,'c2':s2,'c3':s3})

print(df)
```

```python
#loc用index的值進行查詢
print(df.loc['20190604'])

#額外查詢欄位資訊
print(df.loc['20190604':'20190606',['c1','c2']])

#iloc就是用順序查詢
print(df.iloc[1:3,[2,1]])

#查詢欄位
print(df.loc[:,['c1']])
print(df.iloc[:,[1]])
```

```python
# apply: Series.apply(流程)
# ()->執行意思, apply還沒要執行, 只給流程名字

def timeflow(timestamp):
    return str(datetime.utcfromtimestamp(timestamp))
df["film_date_utc"] = df["film_date"].apply(timeflow)
df.head()
```

```python
def tagflow(s):
    tags = eval(s)
    return "children" in tags

df[ df["tags"].apply(tagflow)].head()
```

```python
# ** example
#將整段文字送進去

def test(**a):
    print(a)
    print(a["n"])
    
test(n = 4, b = 9, c="apple")
```

```python
s = input("你想篩選的tag:")
# ** 幫你把參數組合成為字典
# kwargs = {"name":"children"}
def tagflow(s, **kwargs):
    tags = eval(s)
    return kwargs["name"] in tags

df[ df["tags"].apply(tagflow, name=s) ]
```

```python
# T: transpose: 轉置

df.T.to_csv("transpose.csv", encoding="utf-8")
df.T.head()
```
