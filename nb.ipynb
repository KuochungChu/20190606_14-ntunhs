```python
import pandas as pd
traindf = pd.read_csv("poem_train.csv", encoding="utf-8")
traindf.head()
```

```python
testdf = pd.read_csv("poem_test.csv", encoding="utf-8")
testdf.head()
```

```python
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。"
import jieba
" ".join(jieba.cut(s)).replace("\r", "").replace("\n", "")
```

```python
# 1. 分詞, 並且去掉\r\n
def poem(s):
    return " ".join(jieba.cut(s)).replace("\r", "").replace("\n", "")
traindf["內容"] = traindf["內容"].apply(poem)
traindf.head()
```

```python
# 2. 丟掉我不需要的行
traindf = traindf.drop(["詩名"], axis=1)
traindf.head()
```

```python
# 3. 把答案轉化成為0123
trans = {"李白":0, "杜甫":1, "白居易":2}
traindf["作者"] = traindf["作者"].replace(trans)
traindf.head()
```

```python
testdf["內容"] = testdf["內容"].apply(poem)
testdf = testdf.drop(["詩名"], axis=1)
testdf["作者"] = testdf["作者"].replace(trans)
testdf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
xtrain = vec.fit_transform(traindf["內容"])
```

```python
print(xtrain)
```

```python
print(len(vec.vocabulary_))
vec.vocabulary_
```

```python
# 注意: 測試資料不做fit, 只做transform
xtest = vec.transform(testdf["內容"])
print(xtest)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
ytrain = traindf["作者"]
clf.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(xtest)
y_test = testdf["作者"]
print(pre)
print(accuracy_score(pre, y_test))
```

```python
# DEMO
reverse = {0:"李白", 1:"杜甫", 2:"白居易"}
p = input("請輸入一首詩:")
pcut = poem(p)
pvec = vec.transform([pcut])
print("應該是誰寫的?", reverse[clf.predict(pvec)[0]])
```
