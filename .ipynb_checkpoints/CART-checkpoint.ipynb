```python
#應變數: target
#自變數: 除了target

from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["target"] = boston["target"]
df.head()
```

* CRIM：人均犯罪率
* ZN：25,000平方英尺以上民用土地的比例
* INDUS：城镇非零售業商用土地比例
* CHAS：是否鄰近查爾斯河，1是鄰近，0是不鄰近
* NOX：一氧化氮濃度（千万分之一）
* RM：住宅的平均房間數
* AGE：自住且建於1940年前的房屋比例
* DIS：到5個波士頓就業中心的加權距離
* RAD：到高速公路的便捷度指数
* TAX：每萬元的房產税率
* PTRATIO：城镇學生教師比例
* B：1000(Bk − 0.63)2 其中Bk是城镇中黑人比例
* LSTAT：低收入人群比例

```python
#boston["feature_names"]
print(boston["DESCR"])
```

```python
# 訓練集和測試分開來
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(boston["data"], 
                                                    boston["target"], 
                                                    test_size=0.1)
```

```python
# 第一步: fit 模型
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=4)

reg.fit(x_train, y_train)
```

```python
#因為要畫圖，因此引入Graphviz套件
import os
os.environ["PATH"] += os.pathsep + 'C:\\Program Files (x86)\\Graphviz2.38\\bin'

import graphviz
from sklearn.tree import export_graphviz
g = export_graphviz(reg, feature_names=boston["feature_names"], filled=True)
result = graphviz.Source(g)

#存檔boston.pdf
result.render("boston")
result
```

```python
# 第二步: 驗證一下模型好壞

from sklearn.metrics import r2_score
pre = reg.predict(x_test)
print("r2 score:", r2_score(y_test, pre))
```
